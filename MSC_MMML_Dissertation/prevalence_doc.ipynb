# Methodology for Projecting Disease Prevalence and Incidence

This document outlines the mathematical models used to project the prevalence and incidence of Type 2 Diabetes onto a cohort-component population projection for the years 2022–2030. The disease rates are derived from the 2012 study by Sinnott et al., *Estimating the prevalence and incidence of type 2 diabetes.*

A primary data limitation is that the source study provides **total case counts for combined-sex age groups**, but not for sex-specific age groups (e.g., *males aged 15–24*).  
To address this, we developed a method to **disaggregate total case counts into sex-specific counts** by applying the overall male/female proportions of both the disease cases and the general population.

---

## 1. Prevalence Calculation
Prevalence is the total number of existing cases of a disease in a population at a specific point in time.

### 1.1. Derivation of Sex-Specific Prevalence Rates
The sex-specific rates are established once using the 2012 CSO population data from the study’s table and then held constant for the projection period.

#### a) Key Ratios from Source Data (Prevalence Section)

**i. Male vs. Female “Total” Case Ratio**  
- Total Male Cases: $$67,217$$  
- Total Female Cases: $$47,199$$  

Male Proportion of Cases:  
$$ \frac{67,217}{67,217 + 47,199} \approx 58.75\% $$  

Female Proportion of Cases:  
$$ \frac{47,199}{67,217 + 47,199} \approx 41.25\% $$  

**ii. Male vs. Female CSO Population Ratio**  
- Total Male Population: $$2,269,600$$  
- Total Female Population: $$2,315,800$$  
- Total: $$4,585,400$$  

Male Proportion of Population:  
$$ \frac{2,269,600}{4,585,400} \approx 49.5\% $$  

Female Proportion of Population:  
$$ \frac{2,315,800}{4,585,400} \approx 50.5\% $$  

---

#### b) Sex-Specific Rate Calculation for an Age Group ($g$)

1. **Estimate Sex-Specific Cases**  
$$ C_{s,g} = \text{Total Cases}_g \times \text{Sex Proportion of Cases} $$  

2. **Estimate Sex-Specific Population**  
$$ P_{s,g} = \text{Total CSO Population}_g \times \text{Sex Proportion of Population} $$  

3. **Final Sex-Specific Rate**  
$$ R_{\text{prev},s,g} = \frac{C_{s,g}}{P_{s,g}} $$  

---

### 1.2. Final Calculation of Prevalent Cases
For a given year $t$, prevalent cases are calculated as:

$$ \text{Prevalent Cases}_{s,a,t} = P_{s,a,t} \times R_{\text{prev},s,a} $$  

---

## 2. Incidence Calculation
Incidence is the number of new cases of a disease that develop in a population over a specific period.  
The calculation begins from the first projected year (2023).

### 2.1. Derivation of Sex-Specific Incidence Rates
The process mirrors prevalence, but uses the incidence data.

#### a) Key Ratios from Source Data (Incidence Section)

**i. Male vs. Female “Total” Incidence Ratio**  
- Total Male Cases: $$11,980$$  
- Total Female Cases: $$9,201$$  

Male Proportion of New Cases:  
$$ \frac{11,980}{11,980 + 9,201} \approx 56.56\% $$  

Female Proportion of New Cases:  
$$ \frac{9,201}{11,980 + 9,201} \approx 43.44\% $$  

**ii. Male vs. Female CSO Population Ratio (Incidence)**  
- Male Population: $$2,202,383$$  
- Female Population: $$2,268,601$$  
- Total: $$4,470,984$$  

Male Proportion:  
$$ \frac{2,202,383}{4,470,984} \approx 49.26\% $$  

Female Proportion:  
$$ \frac{2,268,601}{4,470,984} \approx 50.74\% $$  

---

#### b) Sex-Specific Incidence Rate Calculation
Using the same formulas as prevalence but substituting incidence data:

$$ R_{\text{inc},s,g} = \frac{C_{s,g}}{P_{s,g}} $$  

---

### 2.2. Final Calculation of Incident Cases

1. **Population at Risk**  
$$ P_{\text{risk},s,a,t} = P_{s,a,t} - \text{Prevalent Cases}_{s,a,t} $$  

2. **Incident Cases**  
$$ C_{\text{inc},s,a,t} = P_{\text{risk},s,a,t} \times R_{\text{inc},s,a} $$  

---

# Methodology for Calculating Sex-Specific Prevalence

To project the number of prevalent cases of Type 2 Diabetes, we apply **age- and sex-specific rates** to projected populations.

The source (2012 data; *Sinnott et al., 2017*) provides:
- **Combined-sex** prevalence by age group (e.g., “15–24 years”)
- **Overall** prevalence for each sex  

but **not** sex-specific rates within each age group. We therefore disaggregate combined-sex rates into sex-specific rates using a **constant male-to-female risk ratio** derived from the overall sex prevalences.

---

## Key Constants (from Table 2)

- Overall male prevalence:  
  $p_{m,all} = 2.96\% = 0.0296$

- Overall female prevalence:  
  $p_{f,all} = 2.04\% = 0.0204$

- Male-to-female risk ratio (assumed age-invariant):  
  $$
  R = \frac{p_{m,all}}{p_{f,all}} = \frac{0.0296}{0.0204} \approx 1.451
  $$
  Thus, on average, men are **~45% more likely** than women to have type 2 diabetes.

---

## Notation (per age group)

- $N_m$: male population in the age group (base year)  
- $N_f$: female population in the age group (base year)  
- $p_{comb}$: combined-sex prevalence for the age group (decimal)  
- $C_{tot}$: total expected prevalent cases in the age group  
- $p_f$: adjusted female-specific prevalence (unknown)  
- $p_m$: adjusted male-specific prevalence (unknown)

---

## Step 1 — Total cases from the combined rate
$$
C_{tot} = (N_m + N_f)\, p_{comb}
$$

---

## Step 2 — Impose constant sex ratio and solve
Assume a constant ratio within the age group:
$$
p_m = R \, p_f
$$

Total cases equal the sum of male and female cases:
$$
C_{tot} = N_m p_m + N_f p_f
        = N_m (R p_f) + N_f p_f
        = (N_m R + N_f)\, p_f
$$

Solve for female prevalence:
$$
p_f = \frac{C_{tot}}{N_m R + N_f}
    = \frac{(N_m + N_f)\, p_{comb}}{N_m R + N_f}
$$

Then male prevalence:
$$
p_m = R\, p_f
    = R \cdot \frac{(N_m + N_f)\, p_{comb}}{N_m R + N_f}
$$

---

## Step 3 — Apply to projected populations (years $t$, 2022–2030)
Let $N_{f,t}$ and $N_{m,t}$ be the projected female and male populations in the age group in year $t$. Then:
$$
\hat{C}_{f,t} = N_{f,t}\, p_f,\quad
\hat{C}_{m,t} = N_{m,t}\, p_m,\quad
\hat{C}_{tot,t} = \hat{C}_{f,t} + \hat{C}_{m,t}
$$
Sum across age groups to obtain annual totals by sex and overall.

---

## Worked Example — Age 15–24 (Base Year 2022)

**Given**
$$
N_m = 330{,}633, \quad
N_f = 319{,}084, \quad
p_{comb} = 0.0013, \quad
R = 1.451
$$

**Step 1: Total cases**
$$
C_{tot} = (330{,}633 + 319{,}084)\times 0.0013
        = 844.6321 \approx 845
$$

**Step 2: Adjusted female rate**
$$
p_f = \frac{844.6321}{330{,}633 \times 1.451 + 319{,}084}
    = \frac{844.6321}{798{,}832.483}
    \approx 0.00105733
$$
Female prevalence $= 0.105733\%$.

**Step 3: Adjusted male rate**
$$
p_m = 1.451 \times 0.00105733 \approx 0.00153419
$$
Male prevalence $= 0.153419\%$.

**Sanity check**
$$
N_m p_m + N_f p_f
= 330{,}633 \times 0.00153419 + 319{,}084 \times 0.00105733
\approx 844.6321 \approx C_{tot}
$$


# Methodology for Calculating Sex-Specific Incidence

To project the number of new (incident) cases of Type 2 Diabetes for each year, we must first determine the "population at risk"—that is, the number of people who do not already have the disease. We then apply an adjusted, sex-specific incidence rate to this at-risk population.

The primary data limitation is that the source study (Sinnott et al., 2017) provides incidence rates for combined-sex age groups (e.g., "15–24 years") but not for sex-specific age groups (e.g., "males aged 15–24"). To address this, we developed a robust method to disaggregate the combined-sex rates into more accurate, sex-specific rates using the overall difference in incidence between men and women as an adjustment factor.

---

## Key Constants from Source Data

The methodology is based on three key constants derived from Table 2 of the study:

- Overall Male Incidence Rate: 0.54 per 1,000 people  
- Overall Female Incidence Rate: 0.41 per 1,000 people  

Male-to-Female Incidence Risk Ratio:

$$
\text{Risk Ratio} = \frac{0.54}{0.41} \approx 1.317
$$

This ratio indicates that, on average, a man is approximately 31.7% more likely to develop the disease in a given year than a woman:

$$
(1.317 - 1) \times 100 = 31.7\%
$$

We assume this risk ratio remains constant across all age groups.

---

## Calculation of Adjusted Sex-Specific Incidence Rates

The adjusted rates are established once using the base year (2022) population data and then held constant for the projection. The formulas are:

**Total Incident Cases in Group (Base Year):**

$$
\text{Total Cases} = (Male\ Population_{2022} + Female\ Population_{2022}) \times Combined\ Rate
$$

**Adjusted Female Incidence Rate:**

$$
\text{Adjusted Female Rate} = \frac{Total\ Cases}{(Male\ Population_{2022} \times Risk\ Ratio) + Female\ Population_{2022}}
$$

**Adjusted Male Incidence Rate:**

$$
\text{Adjusted Male Rate} = \text{Adjusted Female Rate} \times Risk\ Ratio
$$

---

## Final Calculation of New Cases for a Given Year

The number of new incident cases for any given year \(t\) is calculated in two steps, starting from 2023.

**Step 1: Population at Risk**

$$
Population\ at\ Risk_t = Total\ Population_t - Prevalent\ Cases_t
$$

**Step 2: Incident Cases**

$$
Incident\ Cases_t = Population\ at\ Risk_t \times Adjusted\ Incidence\ Rate
$$

---

## Worked Example: 15–24 Age Group (Projection for 2023)

**Step 1: Get Population & Prevalence Data for 2023**

- Male Population (15–24, 2023): 337,162  
- Female Population (15–24, 2023): 325,267  
- Male Prevalent Cases (15–24, 2023): 517  
- Female Prevalent Cases (15–24, 2023): 344  

**Step 2: Get the Pre-Calculated Adjusted Incidence Rates**

- Adjusted Male Incidence Rate: 0.0091% (or 0.000091)  
- Adjusted Female Incidence Rate: 0.0069% (or 0.000069)  

**Step 3: Calculate the Population at Risk for 2023**

$$
Male\ Population\ at\ Risk = 337,162 - 517 = 336,645
$$

$$
Female\ Population\ at\ Risk = 325,267 - 344 = 324,923
$$

**Step 4: Calculate New Incident Cases for 2023**

$$
New\ Male\ Cases = 336,645 \times 0.000091 \approx 31
$$

$$
New\ Female\ Cases = 324,923 \times 0.000069 \approx 22
$$

**Total New Cases:**

$$
31 + 22 = 53
$$
